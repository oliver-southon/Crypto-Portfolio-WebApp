```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("BNB_bars.csv")
df.head()
```

```python
plt.figure(figsize=(12,8))
plt.title("BNB Price Since 2017", fontsize=16)
plt.grid()
plt.plot(df['date'],df[['close']])
plt.tick_params(labelbottom=False)
plt.xlabel("")
```
